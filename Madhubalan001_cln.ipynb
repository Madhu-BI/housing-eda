```python
import pandas as pd
```

```python
df = pd.read_csv(r"D:\blr_house_dataset.csv")
df
```

```python
df.drop('society', axis=1, inplace=True)
```

```python
df.describe()
```

```python
df['Location'] = df['location'].fillna(df['location'].mode()[0])
df.drop('location', axis=1, inplace=True)
```

```python
df['Balcony'] = df['balcony'].fillna(df['balcony'].mode()[0])
df.drop('balcony', axis=1, inplace=True)
```

```python
df['Bathrooms'] = df['bath'].fillna(df['bath'].mode()[0])
df.drop('bath', axis=1, inplace=True)
```

```python
df['Size'] = df['size'].fillna(df['size'].mode()[0])
df.drop('size', axis=1, inplace=True)
```

```python
df.rename(columns={
    'area_type': 'Area_type',
    'availability': 'Availability',
    'total_sqft': 'Total_sqft',
    'price': 'Price₹'
}, inplace=True)
```

```python
df['Size'] = df['Size'].str.extract(r'(\d+)').astype('Int64')

df['Total_sqft'] = df['Total_sqft'].apply(lambda x: (float(x.split('-')[0]) + float(x.split('-')[1])) / 2 if '-' in str(x) else pd.to_numeric(x, errors='coerce')).astype('float')

df['Bathrooms'] = df['Bathrooms'].astype('float')

df['Balcony'] = df['Balcony'].astype('float')

df['Price₹'] = df['Price₹'].astype('int')

df.dropna(subset=['Size', 'Total_sqft', 'Bathrooms', 'Balcony', 'Price₹'], inplace=True)
```
