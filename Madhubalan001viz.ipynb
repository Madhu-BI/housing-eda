```python
import pandas as pd
df = pd.read_csv(r'Madhubalan001_viz.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

custom_colors = ['#3497af', '#8d6041', '#3a2c4f', '#343630']

plt.figure(figsize=(6,4))
sns.countplot(data=df, x='Area_type', hue='Area_type', palette=custom_colors,legend=True)
plt.title("Properties Count by Area Type")
plt.xlabel("Area Type")
plt.ylabel("Count")
plt.xticks(rotation=10)
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.histplot(df['Price'], bins=20, color='#a43165')
plt.title("Price Distribution")
plt.xlabel("Price")
plt.ylabel("Number of Properties")
plt.show()
```

```python
plt.figure(figsize=(6, 4))

custom_colors = ['#cbe50c', '#1ad015', '#0ce5d5', '#e50cd5']


sns.scatterplot(x='Total_sqft', y='Price', data=df, hue='Area_type', palette=custom_colors)
plt.title("Sqft vs Price by Area Type")
plt.xlabel("Total Sqft")
plt.ylabel("Price")
plt.legend(title='Area Type')
plt.show()
```

```python
df.to_csv('Madhubalan001_viz.csv',index =False)
```
